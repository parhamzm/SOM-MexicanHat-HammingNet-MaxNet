```python
import numpy as np
```

```python
input_a = np.array([1, -1, 2, 3, -2])

eps = 0.15
np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
# len(input_a)
```

```python
def AF(x): # Activation Function
    for i in range(0, len(x)):
        if x[i] < 0.0:
            x[i] = 0.0
        else:
            x[i] = x[i]
    return x


def make_MaxNet_Matrix(m=5, eps=0.15):
    maxnet_matrix = np.full((m, m), -1*(eps))
    np.fill_diagonal(maxnet_matrix, 1)
    return maxnet_matrix


def MaxNetDescending(data, eps):
    sorted_array = []
    m = len(data)
    print("Epsilon :=> ", eps)
    print("#################################################")
    a_old = data.copy()
    data1 = data.copy()
    maxnet_mat = make_MaxNet_Matrix(m=m, eps=eps)
    num_non_zero = 0
    i = 0
    for i in range(0, m):
        a_old = data1.copy()
        num_non_zero = 0
        i = 0
        while num_non_zero!=1:
            temp_data = np.inner(maxnet_mat, a_old)
            a_new = AF(temp_data)
            print("Epoch {0}:===> {1}".format(i+1, a_new))
            print("********************************")
            num_non_zero = np.count_nonzero(a_new)
            a_old = a_new
            if num_non_zero == 1 or num_non_zero==0:
                break
            i+=1

        out_tpl = np.nonzero(a_new)
        print("Index of the Max Element :=> ", out_tpl[0].item())
        print("The Value of the Max element is :=> ", data1[out_tpl[0].item()])
        sorted_array.append(data1[out_tpl[0].item()])
        data1[out_tpl[0].item()] = 0
        print("###################################################")
    print("Entered Vector :=> ", data)
    print("Final Sorted Descending Array :=> ", sorted_array)


MaxNetDescending(data=input_a, eps=eps)
```
