```python
import numpy as np
```

```python
input_a = np.array([1.2, 1.1, 1, 0.9, 0.95, 1.15])

eps = 0.15
np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
```

```python
def AF(x): # Activation Function
    for i in range(0, len(x)):
        if x[i] < 0.0:
            x[i] = 0.0
        else:
            x[i] = x[i]
    return x


def make_MaxNet_Matrix(m=6, eps=0.15):
    maxnet_matrix = np.full((m, m), -1*(eps))
    np.fill_diagonal(maxnet_matrix, 1)
    return maxnet_matrix

# a = make_MaxNet_Matrix()
# np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
# print("MaxNet Matrix is :=> ")
# print((a))

def MaxNetAscending(data, eps):
    print("Epsilon :=> ", eps)
    print("#################################################")
    a_old = data.copy()
    data1 = data.copy()
    maxnet_mat = make_MaxNet_Matrix(m=6, eps=eps)
    num_non_zero = 0
    i = 0
    all_zeros = np.array([], dtype=int)
    while num_non_zero!=1:
        temp_data = np.inner(maxnet_mat, a_old)
        a_new = AF(temp_data)
        print("Epoch {0}:===> {1}".format(i+1, a_new))
        print("********************************")
        num_non_zero = np.count_nonzero(a_new)
        # print("Equal Zero :=> ", np.argwhere(a_new == 0))
        zeros = np.argwhere(a_new == 0).reshape(-1)
        zeros_diff = np.setdiff1d(zeros, all_zeros)
        print("Zeros :=> ", zeros)
        print("Zeros Diff : =>", zeros_diff)
        # if all_zeros.size != 0 or i==0:
        all_zeros = np.append(all_zeros, zeros_diff)
        print("All Zeros :=> ", all_zeros)
        a_old = a_new
        if num_non_zero == 1:
            break
        i+=1

    out_tpl = np.nonzero(a_new)
    print("Index of the Max Element :=> ", out_tpl[0].item())
    print("The Value of the Max element is :=> ", data[out_tpl[0].item()])
    all_zeros = np.append(all_zeros, out_tpl[0].item())
    print("All Zeros :=> ", all_zeros)
    print("#####################################")
    print("The Entered Vector :=> ", data)
    print("The Ascending Sorted Vector :=> ", data[all_zeros])


MaxNetAscending(data=input_a, eps=0.09)
```
